```python
def error_libbrecht(T, beta, gamma):
    sig2 = gamma/4/np.pi/T
    fb = np.sqrt(1+beta)*((1+beta)**0.5 + beta**0.5)**3
    return np.sqrt(fb*sig2)
```

```python
T = 1*365*24*3600
gamma = 1*1e-6
beta_list = [0., 0.001, 0.01, 0.1]
```

```python
beta_list = np.logspace(-6, 0, 50)
err = np.array([error_libbrecht(T, beta, gamma) for beta in beta_list])
```

```python
plt.figure()
plt.plot(1/beta_list, err*1e6)
plt.gca().set_xscale('log')
plt.gca().set_yscale('linear')
plt.ylabel('Error (muHz)')
plt.xlabel('S/N')
```
